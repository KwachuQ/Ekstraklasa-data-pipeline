```python
import pandas as pd
import json

# Create the dataframe from JSON response

countries_football = pd.json_normalize(response.json()['categories'])
```

```python
# Create an .md file where there will be a table with name and id of each country
with open('countries_football.md', 'w') as f:
    f.write('| Country Name | Country ID |\n')
    f.write('|--------------|------------|\n')
    for index, row in countries_football.iterrows():
        f.write(f'| {row["name"]} | {row["id"]} |\n')
```
